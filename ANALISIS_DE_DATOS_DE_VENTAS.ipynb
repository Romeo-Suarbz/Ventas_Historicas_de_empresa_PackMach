```python
años = []
for fecha in df['Falta_fac ']:
    años.append(fecha.year)
df["Años"] = años
```

```python
meses = []
for fecha in df['Falta_fac ']:
    meses.append(fecha.month)
df["Meses"] = meses
```

```python
df.head()
```

```python
df.rename (columns = {"Falta_fac " : "Fecha"}, inplace = True)
```

```python
df.rename (columns = {"Años" : "Año"}, inplace = True)
```

```python
df.rename(columns = {"Total_fac           " : "Total_venta"}, inplace = True)
```

```python
df.head()
```

```python
col1 = df["Fecha"]
col2 = df["Total_venta"]
col3 = df["Año"]
col4 = df["Meses"]
col5 = df['Cve_mon     ']
col6 = df['Tip_cam     ']
df_limpio = pd.DataFrame({"Fecha" : col1, "Año" : col3, "Mes": col4, "Total_venta" : col2, "Moneda" : col5, "TC" : col6})
df_limpio
```

```python
df_limpio.loc[df_limpio.Moneda == 1, "TC"] = 1 #DECIMOS QUE CUANDO LA MONEDA SEA LA 1 DENTRO DE NUESTRO DATAFRAME QUE NOS SUSTITUYA EL VALOR DE TC POR 1
```

```python
# df_limpio.loc[df_limpio.Moneda == 2, "Total_MXN"] = df_limpio.loc[df_limpio.Moneda == 2, "Total_venta"] * df_limpio.loc[df_limpio.Moneda == 2, "TC"]
# df_limpio
# AQUI FILTRAMOS EL DATAFRAME DONDE LA COLUMNA DE MONEDA SEA LA 2 Y CREAMOS UNA COL LLAMADA TOTAL_MXN MULTIPLICANDO TOTALVENTA Y TC CUANDO LA MONEDA 
# SEA 2 USAR .LOC SIRVE PARA FILTRAR EL DATAFRAME Y MODIFICAR UNICAMENTE ESA PARTE FILTRADA
```

```python
df_limpio["Total_MXN"] = df_limpio["Total_venta"] * df_limpio["TC"]
df_limpio.columns
```

```python
ventas_por_año = df_limpio.groupby("Año")["Total_MXN"].sum()
```

```python
plt.ticklabel_format(useOffset=False, style='plain')
plt.tick_params(grid_color = "orange", labelcolor = "red", grid_linestyle = "--")
plt.ylabel("Monto Vendido")
plt.xlabel("Año")
plt.title("Ventas Historicas\n")
plt.plot(ventas_por_año, marker = "x", linestyle = ":", color = "green")
plt.axis((2015, 2025, 0, 70000000))
plt.grid()
plt.show()
```

```python
años_top = df_limpio.groupby("Año")["Total_MXN"].sum().sort_values(ascending=False).head(3)
meses_top = df_limpio.groupby("Mes")["Total_MXN"].sum().sort_values(ascending=False).head(3)
tabla = pd.DataFrame({"Años Top" : años_top, "Meses Top" : meses_top})
```
